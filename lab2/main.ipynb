```python
from yargy.tokenizer import MorphTokenizer

tokenizer = MorphTokenizer()

list(tokenizer('сильный'))
```

```python
Action = fact(
    'Action',
    ['force', 'type_force']
)

TYPE_FORCE = morph_pipeline([
    'удар',
    'удары',
    'атака'
])

Force = fact(
    'Force',
    ['kind']
)

KIND_FORCE = or_(
    gram('ADVB'),
    gram('ADJF')
)

FORCE = rule(
    KIND_FORCE.interpretation(
        Force.kind
    )
).interpretation(
    Force
)

ACTION = rule(
    TYPE_FORCE.interpretation(
        Action.type_force.inflected()
    ),
    FORCE.interpretation(
        Action.force
    )
).interpretation(
    Action
)



```

```python
parser = Parser(ACTION)
```

```python
matches = []

for sent in sensts:
    for match in parser.findall(sent):
            matches.append(match.fact)
```
